## Department

#### Number of Departments

```python
dept_count = df['department_id'].nunique()
dept_count
```

```python
dept_list = df['department'].unique().tolist()
dept_list
```

```python
dept_product_counts = df.groupby('department')['product_id'].count().reset_index()
dept_product_counts
```

```python
total_products = df.shape[0]
total_products
```

```python
dept_product_counts['percentage'] = (dept_product_counts['product_id']/total_products)*100
dept_product_counts
```
